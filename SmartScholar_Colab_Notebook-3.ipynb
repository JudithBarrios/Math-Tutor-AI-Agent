This notebook demonstrates an AI agent that performs web research and generates structured summaries using OpenAI and SerpAPI.

```python
# Simple Math Tutor with Built-in Calculator
# Solo Capstone Project by Judith Barrios for ITAI2376

# This function displays the introduction message and instructions to the user.
def display_intro():
    print("-----------------------------------------------------")
    print("Welcome to the Simple Math Tutor!")
    print("Solo Capstone Project by Judith Barrios for ITAI2376.")
    print("-----------------------------------------------------\n")
    print("This tutor helps you practice basic math skills.")
    print("You will be asked one math question at a time.")
    print("Type your answer or type 'hint' for help.")
    print("To enter calculator mode, type 'calc'.")
    print("To end the session, type 'exit', 'quit', or 'end'.\n")
    print("After each question, you'll receive helpful feedback.")
    print("Let's begin!\n")

# This function returns a list of dictionaries,
# each dictionary contains a question, its correct answer, and a hint for the user.
def get_questions():
    return [
        {
            "question": "What is 7 + 5?",
            "answer": 12,
            "hint": "Think of 7, and count forward 5 more: 8, 9, 10, 11, 12."
        },
        {
            "question": "What is 9 - 4?",
            "answer": 5,
            "hint": "Start at 9 and count 4 steps back."
        },
        {
            "question": "What is 6 multiplied by 3?",
            "answer": 18,
            "hint": "Add 6 three times: 6 + 6 + 6."
        },
        {
            "question": "What is 20 divided by 4?",
            "answer": 5,
            "hint": "How many times does 4 go into 20?"
        },
        {
            "question": "What is the remainder when 15 is divided by 4?",
            "answer": 3,
            "hint": "4 fits in 15 three times (4Ã—3=12), remainder is 3."
        }
    ]

# This function asks the user a question and returns their input
def ask_user(question_info):
    print(f"Question: {question_info['question']}")
    return input("Your answer (or type 'hint', 'calc', or 'exit'): ").strip().lower()

# This function prints a hint to help the user solve the current question
def give_hint(hint):
    print(f"Hint: {hint}\n")

# This function checks if the user's answer matches the correct answer
# It tries to convert the input to a number for comparison
def check_answer(user_input, correct_answer):
    try:
        user_number = float(user_input)
        # Allow small difference due to decimal input
        return abs(user_number - correct_answer) < 0.0001
    except ValueError:
        # Return None if input is not a number
        return None

# This function gives detailed feedback based on whether the user's answer was correct or not
def give_feedback(is_correct, user_input, question_info):
    if is_correct:
        print(f"Correct! {user_input} is the right answer.")
        print(f"Explanation: '{question_info['question']}' equals {question_info['answer']}.\n")
    elif is_correct is False:
        print(f"Incorrect. Your answer: {user_input}")
        print(f"The correct answer is: {question_info['answer']}")
        print("Try reviewing how the math operation works and try again.\n")
    else:
        # This happens if the input was invalid (not a number or command)
        print("Invalid input. Please type a number or commands like 'hint', 'calc', or 'exit'.\n")

# This function displays a summary of the session when the user ends the program
def end_session(score, total):
    print("Session ended.")
    print(f"You answered {score} out of {total} questions correctly.")
    print("Thanks for practicing math with the tutor. Goodbye!")

# Calculator mode lets the user type in math expressions to be calculated by the program
def calculator_mode():
    print("\n--- Calculator Mode ---")
    print("Type a math expression (e.g., 4 + 7 * 2) or 'back' to return to questions.")
    while True:
        expr = input("Calc> ").strip().lower()
        # Allow the user to exit calculator mode
        if expr in ['back', 'exit', 'quit']:
            print("Exiting calculator mode.\n")
            break
        try:
            # Only allow safe characters in the expression
            allowed = "0123456789+-*/(). "
            if all(c in allowed for c in expr):
                # Evaluate and print the result
                result = eval(expr)
                print(f"Result: {result}\n")
            else:
                # Warn if invalid characters are used
                print("Invalid characters in expression. Only numbers and + - * / ( ) are allowed.\n")
        except Exception:
            # Catch errors like malformed expressions and prompt again
            print("Error evaluating expression. Please enter a valid math expression.\n")

# Main function that runs the math tutor program
def run_tutor():
    display_intro()  # Show introduction and instructions
    questions = get_questions()  # Get all questions
    correct_answers = 0  # Track number of correct answers
    total_questions = 0  # Track total answered questions
    index = 0  # Current question index

    while True:
        current_question = questions[index]
        user_input = ask_user(current_question)

        # Check if user wants to exit the program
        if user_input in ['exit', 'quit', 'end']:
            end_session(correct_answers, total_questions)
            break
        # If user wants a hint, provide it without counting the question as answered
        elif user_input == 'hint':
            give_hint(current_question['hint'])
            continue
        # If user wants to use the calculator, enter calculator mode
        elif user_input == 'calc':
            calculator_mode()
            continue
        else:
            # Check the answer and give feedback
            result = check_answer(user_input, current_question['answer'])
            give_feedback(result, user_input, current_question)
            # Only count the question if the input was a valid number answer
            if result is not None:
                total_questions += 1
                if result:
                    correct_answers += 1
        # Move to the next question, loop back to the start when done
        index = (index + 1) % len(questions)

# Entry point of the program; runs the tutor if this script is executed
if __name__ == "__main__":
    run_tutor()
```
